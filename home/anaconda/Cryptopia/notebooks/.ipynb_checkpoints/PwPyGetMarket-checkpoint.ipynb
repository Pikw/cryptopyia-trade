```python
import requests
import json
import pandas as pd
```

```python
path = 'https://www.cryptopia.co.nz/api/GetMarketHistory/DOT_BTC/48'
# id ZEC = 4327
r = requests.get(path)
r
# r.text.encode('utf-8')
```

```python
jsload = json.loads(r.text)#
jsloadcut = jsload['Data']
# jsload['Data']
```

```python
from pandas.io.json import json_normalize
df = json_normalize(jsload['Data'])
```
